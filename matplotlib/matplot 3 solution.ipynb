# Pie Charts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
pokemon=pd.read_csv('pokemon.csv')
print("shape =",pokemon.shape)
pokemon.head(10)
```

```python
pkmn_type1=pokemon[['id','species','type_1']]
pkmn_type1.head()
```

```python
# by using pandas
ord1=pkmn_type1['type_1'].value_counts().index
print("orderd types:\n",ord1)
print()
counts=pkmn_type1['type_1'].value_counts()
print("count for every type:")
counts
```

```python
plt.pie(counts, labels = ord1, startangle = 45,counterclock = False);
plt.axis('square')
```

```python
total_pokemon=pokemon.shape[0]#807
```

```python
ord1
```

```python
ord_lst=list()
for name in ord1:
    ord_lst.append(name+"\n"+ str(int(counts[name]/total_pokemon*100))+"%")#"water\n14%"



ord_lst    
```

```python
plt.pie(counts, labels = ord_lst, startangle = 45,counterclock = False);
plt.axis('square')
```

```python
plt.pie(counts, labels = ord1, startangle = 90,counterclock = False, wedgeprops = {'width' : 0.8});
plt.axis('square')
```
